## 기타 분류

```python
#판매량이 가장 높은 상품 분류, 조인 사용
df_product = pd.merge(df2, df4, left_on ='pd_c', right_on = 'pd_c', how='inner')
```

```python
df_product
```

```python
product_sum = df_product.groupby(['clac_hlv_nm', 'clac_mcls_nm']).sum()['buy_ct']
product_sum
```

```python
top_sailed_category = product_sum.sort_values(ascending = False).head(10)
top_sailed_category
```

```python
top_sailed_category.plot.barh("가장 많이 판매된 카테고리 상품")
```

```python
#가장 많이 판매된 품목 알아보기
top_sailed_product = df_product.groupby(by = 'pd_nm').sum()['buy_ct']
```

```python
top_sailed_product = top_sailed_product.sort_values(ascending = False).head(10)
```

```python
top_sailed_product.plot.pie()
```

```python
plt.figure(figsize = (20, 10))
top_sailed_product.plot.barh(title = "가장 많이 판매된 상품", fontsize =20)
```
